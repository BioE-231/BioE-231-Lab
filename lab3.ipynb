```python
import matplotlib.pyplot as plt
%matplotlib inline
from math import log10

romeo = open("romeoandjuliet.txt")
lines = romeo.read().split() # Generate a list where elements are words in the text.

newlines = [] # Generate a new list which unifies all the words into lowercase and eliminates common punctuations.
for i in lines:
    newlines.append(i.lower().strip(",.?]['!;:-)("))

names = set(newlines)

dict = {} # Generate a new dictionary where keys are the unique words and values are the times they show up.
for name in names:
    count = 0
    for name1 in newlines:
        if name == name1:
            count += 1
    dict[name] = count
    
frequency = sorted(dict.items(), key = lambda x: x[1], reverse = True) # Sort the dict with the decreasing word count. Return tuples.
```

```python
Most_ten = []
Most_ten_value = []
for i in range(10):
    Most_ten.append(frequency[i][0])
    Most_ten_value.append(frequency[i][1])

Least = []
for i in frequency:
    if i[1] == 1:
        Least.append(i[0])

print("The most ten words are: ", Most_ten)
print('\n')
print("The least words are: ", Least)
```

### The most ten words and the least words.
As you can see above, the most frequent word in *Romeo and Juilet* are those prepositions, personal pronoun and articles. It matches the common sense and English writing habit.

```python
plt.bar(x = Most_ten, height = Most_ten_value)
plt.title('Top 10 most frequent words')
plt.xlabel('Top 10 most frequent words')
plt.ylabel('Word_Count')
plt.show()
```

### The word_count difference between top ten.
As the **Bar plot** shown above, the difference between the top 6 words is relatively larger than the rest. This reflects the tendency that the less the words are shown, the more words are of the relatively same frequency.

```python
values = [] # Create a list contains the decreasing word count.
for i in frequency:
    values.append(i[1])
    
rank_count = plt.plot(range(1, len(dict)+1), values)
plt.xscale('log')
plt.yscale('log')
plt.title('Word Count vs. Rank (log)')
plt.xlabel('Rank of the Word (Log 10)')
plt.ylabel('Word_Count (Log 10)')
plt.show()

plt.plot(range(1, len(dict)+1), values)
plt.title('Word Count vs. Rank')
plt.xlabel('Rank of the Word')
plt.ylabel('Word_Count')
plt.show()
```

### The Word count vs. Rank
The above figures show the difference between the figures using different scale of axis. Though they look totally distinct; however, they are actually using the same data. Compared with linear scale of axis, for those data with enormous range, I prefer to use Log scale instead for it shows more details when x-axis is small.
